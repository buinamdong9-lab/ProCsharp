# [dbo].[sp_ConfirmReturn]

```python
/****** Object:  StoredProcedure [dbo].[sp_ConfirmReturn]    Script Date: 4/20/2026 8:30:52 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- 4) Xác nhận trả
CREATE PROCEDURE [dbo].[sp_ConfirmReturn]
    @TicketID INT,
    @ReturnedBy INT,
    @ReturnNote NVARCHAR(255) = NULL
AS
BEGIN
    SET NOCOUNT ON;

    UPDATE dbo.BorrowTickets
    SET Status = N'RETURNED',
        ReturnDate = GETDATE(),
        ReturnedBy = @ReturnedBy,
        Note = ISNULL(Note, N'') + CASE WHEN @ReturnNote IS NULL THEN N'' ELSE N' | ' + @ReturnNote END
    WHERE TicketID = @TicketID
      AND Status = N'BORROWING';
END
GO
```

# [dbo].[sp_CreateBorrowTicket]

```python
/****** Object:  StoredProcedure [dbo].[sp_CreateBorrowTicket]    Script Date: 4/20/2026 8:30:52 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- 2) Tạo phiếu mượn (tạo header)
CREATE PROCEDURE [dbo].[sp_CreateBorrowTicket]
    @UserID INT,
    @CreatedBy INT = NULL,
    @ExpectedReturnDate DATETIME = NULL,
    @Note NVARCHAR(255) = NULL
AS
BEGIN
    SET NOCOUNT ON;

    INSERT INTO dbo.BorrowTickets(UserID, CreatedBy, ExpectedReturnDate, Note)
    VALUES(@UserID, @CreatedBy, @ExpectedReturnDate, @Note);

    SELECT SCOPE_IDENTITY() AS NewTicketID;
END
GO
```

# [dbo].[sp_Login]

```python
/****** Object:  StoredProcedure [dbo].[sp_Login]    Script Date: 4/20/2026 8:30:52 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- 1) Login (đơn giản: demo theo Username/PasswordHash)
CREATE PROCEDURE [dbo].[sp_Login]
    @Username NVARCHAR(50),
    @PasswordHash NVARCHAR(255)
AS
BEGIN
    SET NOCOUNT ON;

    SELECT u.UserID, u.Username, u.FullName, u.RoleID, r.RoleName
    FROM dbo.Users u
    JOIN dbo.Roles r ON r.RoleID = u.RoleID
    WHERE u.Username = @Username
      AND u.PasswordHash = @PasswordHash
      AND u.IsLocked = 0;
END
GO
```
